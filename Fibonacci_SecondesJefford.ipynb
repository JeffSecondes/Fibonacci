```python
## Submitted by: Jefford Secondes ###

import time                           ###  importing two libraries, time and matplotlib.pyplot, 
import matplotlib.pyplot as plt

def fibonacci_recursion(n):   ### we have a function called fibonacci_recursion, takes in an input of n and calculates the nth Fibonacci number using recursion.
    if n <= 0:              #If n is 0 or less, it returns 0.
        return 0    
    elif n == 1:            #If n is 1, it returns 1.
        return 1
    else:                   # If n is more than 1, it returns the sum of the (n-1)th and (n-2)th Fibonacci numbers.
        return fibonacci_recursion(n-1) + fibonacci_recursion(n-2)

start = time.time()         # we use the time library to measure how long it takes to calculate the 10th Fibonacci number
result = fibonacci_recursion(10)    #We then run fibonacci_recursion(10) and store the result in result.
end = time.time()           #save the current time in end.
duration = end-start        #subtract start from end and store the result in duration, represents how long it took to calculate the 10th Fibonacci number.
print("Fibonacci number:", result)  # we print the result and the duration.
print("Duration:", duration, "seconds") #  we print the result and the duration.

x_axis_output_result = [] #creates two empty lists, x_axis_output_result and y_axis_duration. #It sets a variable input_user to 30.
y_axis_duration = []
input_user = 30

for i in range(0,input_user):  #for loop that goes from 0 to input_user, and runs the following steps for each number i:
    start = time.time()     #Measure the start time.
    result = fibonacci_recursion(i)  # Calculate the ith Fibonacci number using the fibonacci_recursion function and store the result in result.
    end = time.time()    # Measure the end time
    duration = end-start  #Calculate the duration as end minus start and store the result in duration.

    x_axis_output_result.append(result)    #Append the result to x_axis_output_result.
    y_axis_duration.append(duration)  #Append the duration to y_axis_duration.

print(x_axis_output_result)
print(y_axis_duration[i])


plt.plot((x_axis_output_result), y_axis_duration,  "o")  ## The x_axis_output_result is the x-axis, y_axis_duration is the y-axis, and the "o" is the marker.
plt.xlabel("result")   #sets the x-axis label as "result", y-axis label as "time", and the title as "Fibonacci sequence (recursion)
plt.ylabel("time")
plt.title("Fibonacci sequence (recursion)")
plt.show()    # it shows the plot
 
```

The Fibonacci sequence can be determined by utilizing either a recursive or a non-recursive approach to the calculation. The recursive approach requires making many calls to the same function, and as a result, it consumes more memory and takes more time overall, particularly when the sequence reaches higher values. On the other hand, the non-recursive approach makes use of iteration rather than recursion and, as a result, requires less memory and can process big values of the sequence more quickly. On the other hand, in comparison to the straightforward recursive technique, the non-recursive approach could be more difficult to comprehend and put into practice. Both of these options come with certain drawbacks, and which one is preferable is contingent on the details of the job at hand.
